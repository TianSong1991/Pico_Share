## Python高阶进修概念

<font face="微软雅黑" size=4>1、闭包：在一个外函数中定义了一个内函数，内函数里运用了外函数的临时变量，并且外函数的返回值是内函数的引用。这样就构成了一个闭包。<br>
2、lambda表达式：无头表达式，无名函数。<br>
3、装饰器：修改其他函数的功能的函数，函数套函数进行调用使用。<br>
4、迭代器：可以被next（）调用并不断返回下一个值得对象称为迭代器（Iterator）。<br>
5、生成器：与yield对应，只要把一个列表生成式的[]改成()，就创建了一个generator。 如果一个函数中包含yield关键字，那么这个函数就不再是一个普通函数，而是一个generator。<br>
6、纯函数：没有副作用的函数成为纯函数，即没有其他变量参数的影响。<br>
7、map，reduce、filter函数<br>
8、数字，布尔，字符串，列表，元组，字典<br>
9、可变类型与不可变类型：这里的可变不可变，是指内存中的那块内容（value）是否可以被改变

## Depth、IR、RGB图像

<font face="微软雅黑" size=4>Depth图：深度图单通道，存储着深度值，即物体到相机的欧式距离。深度图可转换为点云图，通过cloudcompare软件打开查看。<br>
IR图：单通道图，可以类似的理解为灰度图d，但是与灰度图有区别，区别在于IR接受外界的能量生成的图像。<br>
RGB图：三通道图，即平常看的彩色图。

## 8位、16位、32位图

### 8位图

<font face="微软雅黑" size=4>8位的意思是说，能够显示出来的颜色的数量有8位数。16、24都是一样的。<br>
256色模式： 8 bits (B: 2 bits; G: 3 bits; R: 3 bits) <br>
缺点：颜色亮度少 颜色少<br>
优点：图片占用内存小数据传输快<br>

### 16位图

<font face="微软雅黑" size=4>16 bits （B: 5 bits, G: 6 bits, R: 5 bits)<br>
16位颜色值，可心这样分：X1位，R5位，G5位，B5位；<br>
R5位，G6位，B5位或者是X、R、G、B均4位等等；<br>
说明：在颜色数方面，当然很明显，是2的16次方、24次方和32次方的差别。<br>
从人眼的感觉来说，16位色能基本满足显示需要，粗看起来和24位色、32位色差不多，<br>
仔细点研究的话，会发现在大面积的渐变色中16位色显示会出现隐约的分隔线。<br>

### 24位图

<font face="微软雅黑" size=4>24位是R、G、B三个通道各占6位共18位，加上明度通道6位，所以是24位<br>
24位颜色值称为“真彩色”，它有着统一的颜色分配：R8位，G8位，B8位；<br>
说明：实际应用中，最常用的是24位色，比如网页和其他地方表示颜色的 “#80FF2E”之类的字符串，<br>
用答CDSee看图片也可以看到大部分图片的色深是24位。<br>
优点：图片占用内存小数据传输快<br>
缺点：24bit 和 32bit 是一样的，由于24bit不能在显存中对齐，<br>

### 32位图

<font face="微软雅黑" size=4>32位是R、G、B三个通道各占8位共24位，加上透明度8位，所以是32位<br>
32位颜色值一般这样分配：X8位，R8位，G8位，B8位或A8位，R8位，G8位，B8位。<br>
说明：至于32位色，那是在24位基础上加出来的，据说是加了8位透明色，<br>
实际和24位色没多大差别，至少我们的眼睛是分辨不出来的。<br>
优点：32bit 因为可对齐，大多数的显卡都支持<br>
缺点：占用显存较大<br>
至于32位色，那是在24位基础上加出来的，据说是加了8位透明色，<br>
实际和24位色没多大差别，至少我们的眼睛是分辨不出来的。<br>
